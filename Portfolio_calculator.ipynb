```python
shares = 10, 2 #how many you own
prices = 150.25, 800.50 #current prices

total = 0
for i in range(len(shares)):
    value = shares[i] * prices[i]
    total += value
print(f"total portfolio value: ₹{total}")
```

```python
prices = []
shares = []
total = 0
for i in range(3):
    p = float(input(f"Enter price for stock {i+1}: "))
    s = int(input(f"Enter shares for stock {i+1}: "))
    prices.append(p)
    shares.append(s)

print("\n---portfoio breakdown---")
for i in range(len(shares)):
    value = shares[i] * prices[i]
    print(f"share{i+1} * ₹{prices[i]} = ₹{value}")
    total += value
print(f"\ntotal portfolio value: ₹{total}")
```
